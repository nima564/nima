```python
#ابندا یک تابع یا فانکشن با دو متغییر با آرگومان تعریف میکنیم و بعد یک شرط برای بررسی به رسیدن هدف خود مینویسیم و بعد از ان دستوراتی را مینویسیم برای درست نبودن شرط  
def CSGO(z,x):
   
    if(z==0):
        print("done")
        print("")
        print("0")
        print("")
        print("gameover")
       #اسمبر تا اخر کد ها را میخون و مواقعی که شرط درست نیست میره دستوراتی که ما در درست نبودن شرط نویشتم رامیخونه و همین را تکرار میکنه تا شرط ما درست بشه وما با اینکار یک حلقه ای درست کردیم
        
    else:
        print(z)
        print("")
        print("-")
        print("")
        print(x)
        print("")
        print("=")
        print("")
        print(z-x)
        print("")
        print("//////////////")
        print("")
        #اگر ما اگر ما دستور زیر را ننویسیم یک حلقه بینهایت درست کردیم و هیچوقت خاتم نمی یابد
        #دستور ما کاری که میکند این است که دوعدد را از ارگومان میگیرد و با دستور زیر این دو عدد را کم میکند و حاصل کم کردن اعداد را نشان میدهد
        
        CSGO(z-x,x)
        #اگر ما اعدادی که برای ارگومان تعریف میکنیم بخش پذیر نباشد حلقه ای ایجاد میکند که خاتم نمی یابد 
CSGO(18,3)   
```
